## Ejercicio 4 (1.5)

Queremos ahora proteger los datos en `df0` y para ello vamos a aplicar ruido aditivo no correlacionado a todos los atributos (todas las notas).

Define una función llamada `additive_noise` que reciba como parámetro un
`DataFrame` como `df0`, el parámetro *p* que determina el nivel de ruido
y retorne un `DataFrame` nuevo con el ruido añadido.

El ruido aditivo, se obtiene de forma aleatoria de una distribución normal
$N(\mu, \sigma^2)$ con una media $\mu=0$ y una varianza
$\sigma^2=p\sigma^2_a$, donde $\sigma^2_a$ es la varianza de los valores
originales.

*Hint:* [`numpy.random.normal`](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)

```python
def additive_noise(DataFrame, p):
  # Determinamos mu como 0 y sigma como la varianza 
  mu, sigma = 0, p
  num_rows, num_columns = DataFrame.shape
  
  # Calculamos el ruido mediante los parametros mu y sigma para las dimensiones
  # del DataFrame pasado como parametro
  noise = np.random.normal(mu, sigma, [num_rows, num_columns])

  return DataFrame + noise
```

## Ejercicio 7 (2)

Aunque el ataque anterior no tiene éxito podemos intentar aproximarlo. Puede ser que no haya p.e. ningún registro con `subject0 == 7.7` pero puede ser que tenga un valor muy cercano.

Para ello podemos buscar los registros en `df1` que cumplan las siguientes condiciones:
- $7.6 < subject0 < 7.8$
- $6.5 < subject1 < 6.7$

Dado que el proceso de anonimización con ruido tiene un alto grado de aleatoriedad, vamos a repetir 5 veces las siguientes operaciones:
1. Aplicar `additive_noise` con $p=0.01$ a `df0`
2. Seleccionar los registros que cumplan las condiciones anteriores sobre `subject0` y `subject1`.
3. Mostrar el número de registros que cumplen la condicion.
4. Mostrar si está el registro que buscamos entre los registros seleccionados.

Como resultado tenéis que mostrar el número de registros seleccionados, y si el registro que buscamos está en los seleccionados para las 5 ejecuciones.

*Hint*: se puede identificar el registro que buscamos por su índice (el índice del registro obtenido en el Ejercicio 3.


```python
# Determinamos los parametros que seran utilizados por nuestra funcion de busqueda
p, reps = 0.01, 5
subject0_low, subject0_high = 7.6, 7.8
subject1_low, subject1_high = 6.5, 6.7
target_student = 6021

"""
Denominamos asi a la funcion de busqueda en base a unos intervalos de calificaciones
para las asignaturas subject0 y 1.
"""
def finding_nemo(DataFrame, p, reps, verbose):
  # En esta lista guardaremos los resultados de las DataFrames modificadas
  # asi, estas podran ser utilizadas por sucesivas funciones creadas en
  # apartados siguientes
  noisy_df_list = []

  for i in range(reps):
    # Aplicamos la funcion additive noise a la DataFrame en base a la varianza definida
    noisy_df = additive_noise(DataFrame, p)
    noisy_df_list.append(noisy_df)
    # Si se encuentra una fila con las asingaturas 0 y 1 en el rango asignado
    # se almacena en la dataframe matches_found
    matches_found = noisy_df.loc[(noisy_df['subject0'].between(subject0_low, subject0_high)) & \
                 (noisy_df['subject1'].between(subject1_low, subject1_high))]
    # En caso de especificar verbose como True, se imprimen los resultados de las busquedas
    if verbose: 
      print('Found {} rows for iteration {}\n'.format(matches_found.shape[0], i))
      print(matches_found)
      print('Student 6021 found: {} \n\n'.format(target_student in matches_found.index))

  return noisy_df_list

df_list = finding_nemo(df0, p, reps, True)
```

**Resultado**: se ha econtrado el alumno 6021 en las 5 iteraciones

## Ejercicio 9 (1)

Al proteger los datos en `df1` estamos introduciendo una perdida de información. Para medir esta perdida vamos a utilizar el *Mean Squared Error (MSE)* tal como se define en los materiales de la asignatura.

Define una funcion `mse` que dados dos `DataFrame`s retorne el MSE entre ellos, y repite el experimento anterior de las 5 ejecuciones mostrando en cada caso el MSE

```python
"""
Mean Squared Error (MSE): mide la media de la diferencia elemento a
elemento al cuadrado. Una característica de esta medida es que da más
peso a valores atípicos (outliers) al elevar al cuadrado la diferencia.
MSE(A,B) = 1/NM * SUMij (aij - bij)^2
"""
def mse(original_df, noisy_df):
  return np.mean((original_df - noisy_df) ** 2)
```

```python
# Experimentation cell
noisy_df_list = finding_nemo(df0, p, reps, False)
for noisy_df in noisy_df_list:
  print(mse(df0, noisy_df))
  print('\n')
```
